```python
# Importing modules for the program
import tkinter as tk
import datetime
class Countdown(tk.Frame):
    def __init__(self, master):
        super().__init__(master)
        self.create_widgets()
        self.show_widgets()
        self.seconds_left = 0
        self._timer_on = False

    # For displaying the windows
    def show_widgets(self):
        self.Label.pack()
        self.entry.pack()
        self.start.pack(side=tk.TOP)
        self.stop.pack(side=tk.LEFT,fill="y")
        self.pause.pack(side=tk.TOP)
        self.resume.pack(side=tk.TOP)
        self.reset.pack(side=tk.BOTTOM,fill="y")

    # For creating the widgets in the window
    def create_widgets(self):
        self.Label = tk.Label(self, text="Enter the time")
        self.entry = tk.Entry(self, justify="center")
        self.entry.focus_set()
        self.reset = tk.Button(self, text="Reset", command=self.reset_button)
        self.stop = tk.Button(self, text="Stop", command=self.stop_button)
        self.start = tk.Button(self, text="Start", command=self.start_button)
        self.pause = tk.Button(self, text="Pause", command=self.pause_button)
        self.resume = tk.Button(self, text="Resume", command=self.resume_button)

    def convert_seconds_left_to_time(self):
        return datetime.timedelta(seconds=self.seconds_left)

    def countdown(self):
        self.Label["text"] = self.convert_seconds_left_to_time()

        if self.seconds_left:
            self.seconds_left -= 1
            self._timer_on = self.after(1000, self.countdown)

        else:
            self.timer_on = False

    # Functions of the countdown timer
    # They are in the form of buttons 
    def reset_button(self):
        self.seconds_left = int(self.entry.get())
        self.stop_timer()
        self._timer_on = False
        self.label["text"] = "Enter the time in seconds:"
        self.start.forget()
        self.stop.forget()
        self.reset.forget()
        self.pause.forget()
        self.resume.forget()
        self.start.pack(side=tk.TOP)
        self.stop.pack(side=tk.LEFT,fill="y")
        self.reset.pack(side=tk.BOTTOM,fill="y")
        self.pause.pack(side=tk.TOP)
        self.resume.pack(side=tk.TOP)

    def stop_timer(self):
        if self._timer_on:
            self.after_cancel(self._timer_on)
            self._timer_on = False

    def stop_button(self):
        self.seconds_left = int(self.entry.get())
        self.stop_timer()

    def start_button(self):
        self.seconds.left = int(self.entry.get())
        self.start_timer()
        self.countdown()
        self.start.forget()
        self.stop.forget()
        self.reset.forget()
        self.pause.forget()
        self.resume.forget()
        self.start.pack(side=tk.TOP)
        self.stop.pack(side=tk.LEFT,fill="y")
        self.reset.pack(side=tk.BOTTOM,fill="y")
        self.pause.pack(side=tk.TOP)
        self.resume.pack(side=tk.TOP)

    def resume_countdown(self):
        self.Label["text"] = datetime.timedelta(seconds=self.seconds_left)
        if self.seconds_left:
            self.seconds_left -= 1
            self.timer_on = self.after(1000, self.countdown)

        else:
            self._timer_on = False

    def resume_button(self):
        print("Time left when resuming", self.seconds_left)
        self.resume_countdown()

    def pause_button(self):
        self.stop_timer()
        self._timer_on = True
        print("Time left when pausing:", self.seconds_left)

    if __name__ == "__main__":
        print("r")
        root = tk.Tk()
        root.geometry("500x200")
        root.resizable(True, True)
        root.title("Countdown timer")
        root.configure(highlightbackground="white", background="white", bg="white")
        countdown = Countdown(root)
        countdown.pack()
        countdown.configure(highlightbackground="blue", background="blue", bg="blue")

        root.mainloop()
```
